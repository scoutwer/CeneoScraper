### Import pakietów 

```python
import os 
import pandas as pd 
import numpy as np 
from matplotlib import pyplot as plt
```

### Wczytanie i przygotowanie danych

```python
print(*[filename.split(".")[0] for filename in os.listdir("./opinions")], sep="\n")
```

```python
product_id = input("Podaj kod produktu: ")
```

```python
opinions = pd.read_json(f"./opinions/{product_id}.json")
```

```python
opinions.stars = opinions.stars.apply(lambda s: s.split("/")[0].replace(",",".")).astype(float)
opinions.useful = opinions.useful.astype(int)
opinions.useless = opinions.useless.astype(int)
opinions
```

### Podstawowe statystyki

```python
opinions_count = opinions.shape[0]
pros_count = opinions.pros.astype(bool).sum()
cons_count = opinions.cons.astype(bool).sum()
pros_cons_count = opinions.apply(lambda o: bool(o.pros) and bool(o.cons), axis=1).sum()
avg_stars = opinions.stars.mean()
print(f"Liczba opini o produkcie: {opinions_count}")
print(f"Liczba opini z podaną  listą zalet: {pros_count}")
print(f"Liczba opini z podaną  listą wad: {cons_count}")
print(f"Liczba opini z podaną  listą zalet i listę wad: {pros_cons_count}")
print(f"Średnia ocena produktu: {avg_stars:.3}")
```

```python
pros = opinions.pros.explode().dropna().value_counts()
cons = opinions.cons.explode().dropna().value_counts()
print(f"Lista unikatowych zalet: {len(pros)}")
print(f"Lista unikatowych zalet: {len(cons)}")
print(pros, cons)
```

### Wykresy

```python
recomendations = opinions.recommendation.value_counts(dropna=False).reindex(["Nie polecam", "Polecam", None])
recomendations
```

```python
recomendations.plot.pie(label="", title = f"Rozkład rekomendacji w opiniach o produkcie {product_id}", 
                        colors = ['crimson','forestgreen', "lightGrey"],
                        labels = ["Nie polecam", "Polecam", "Nie mam zdania"],
                        autopct = "%1.1f%%")
```
